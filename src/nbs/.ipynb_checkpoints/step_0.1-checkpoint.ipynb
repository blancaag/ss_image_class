```python
def plot_images(data, ncols=3):
    ncols = ncols 
    nrows = data.shape[0] // ncols
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols)
    fig.set_size_inches(12, nrows*5)
    fig.tight_layout()
    for i, ax in enumerate(axes.flat):
        im = ax.imshow(data[i])
        ax.set_axis_off()
        ax.title.set_visible(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        for spine in ax.spines.values():
            spine.set_visible(False)
    plt.subplots_adjust(left=0, hspace=0, wspace=0)
    plt.show()
    
plot_images(x_train[:3])
```

## Training

#### Simple CNN

```python
output_folder = 'simple_cnn'
if not os.path.exists(output_folder): os.mkdir(output_folder)

# model arquitecture

m = Sequential([
        BatchNormalization(axis=1, input_shape=input_shape),
        Conv2D(32, 3, activation='relu'),
        BatchNormalization(),
        MaxPooling2D(),
        Conv2D(32, 3, activation='relu'),
        BatchNormalization(),
        MaxPooling2D(),
        Conv2D(32, 3, activation='relu'),
        BatchNormalization(),
        MaxPooling2D(),
#         MaxPooling2D(),
        #
        Conv2D(32, 3, activation='relu'),
        BatchNormalization(),
        MaxPooling2D(),
        Conv2D(32, 3, activation='relu'),
        BatchNormalization(),
        MaxPooling2D(),
#         MaxPooling2D(),
        #
        Flatten(),
        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(1, activation='sigmoid')
    ])

m.summary()

m.compile(Adam(), loss='binary_crossentropy', 
          metrics=['accuracy', recall, precision, fmeasure])

# m.load_weights(output_folder + '/simple_cnn_w.h5')
```

```python
m.optimizer.lr = 0.0001; print('lr: ', m.optimizer.lr)
aug_data_gen = gen_
epochs = 10
batch_size = 128

gen_train = aug_data_gen.flow(x_train, y_train, batch_size=batch_size)
gen_val = data_gen.flow(x_test, y_test, batch_size=batch_size)

print(gen_train.n, gen_val.n)

start = time.time()
h = m.fit_generator(gen_train, gen_train.n//gen_train.batch_size, 
                    epochs=epochs, workers=10,
                    validation_data=gen_val, validation_steps=gen_val.n//gen_val.batch_size)

end = time.time()
print(end-start)
```

```python
m.save_weights(output_folder + '/simple_cnn.h5')
```

```python
score = m.evaluate(x_train, y_train, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
print('Test binary_accuracy:', score[2])
print('Test categorical_accuracy:', score[3])
print('Test recall:', score[4])
print('Test precision:', score[5])
```
